```python
# ⚙️ Pré-configurar UI com LoRA, resolução 1024x1024, batch=1 e prompt negativo completo
import json
ui_config = {
    "txt2img_prompt": "<lora:zzzzzzz:0.7>",
    "txt2img_negative_prompt": "censored, lowres, (worst quality, bad quality:1.2), bad anatomy, sketch, jpeg artifacts, signature, watermark, old, oldest, chibi, loli, simple background, missing fingers, pubic hair, extra digit, fewer digits, extra fingers,mutated, deformed, long body, bad anatomy, mosaic censoring, bar censor, deformed face, asymmetrical eyes, blurry face, fused eyes, bad anatomy, multiple pupils, cross-eyed, duplicate eyes, mutated face, glitched eyes, extra limbs, low resolution, worst quality",
    "txt2img_width": 1024,
    "txt2img_height": 1024,
    "txt2img_batch_size": 1,
    "txt2img_cfg": 7,
    "txt2img_steps": 25,
    "txt2img_sampler": "Euler a",
    "txt2img_enable_hr": True,
    "txt2img_hr_upscaler": "R-ESRGAN 4x+ Anime6B",
    "txt2img_hr_scale": 1.35,
    "txt2img_denoising_strength": 0.4
}
with open("/content/sd-webui/ui-config.json", "w") as f:
    json.dump(ui_config, f, indent=2)
print("Configuração da interface aplicada com sucesso.")
```
